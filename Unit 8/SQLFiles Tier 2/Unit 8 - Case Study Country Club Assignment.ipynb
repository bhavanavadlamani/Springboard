### Q7: Produce a list of all members who have used a tennis court. Include in your output the name of the court, and the name of the member formatted as a single column. Ensure no duplicate data, and order by the member name.

```python
df_mem['fullname'] = df_mem['firstname'].astype(str) + df_mem['surname'].astype(str)
```

```python
ps.sqldf('SELECT DISTINCT fullname AS mem_used_tennis_court, name as name_of_court FROM (SELECT mem.*, fac.facid ,fac.name ,bk.* FROM df_bk AS bk LEFT JOIN df_fac AS fac ON fac.facid = bk.facid LEFT JOIN df_mem AS mem ON bk.memid = mem.memid WHERE fac.name LIKE "Tennis%") WHERE memid <> 0 ORDER BY fullname')
```

### Q8: Produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30. Remember that guests have different costs to members (the listed costs are per half-hour 'slot'), and the guest user's ID is always 0. Include in your output the name of the facility, the name of the member formatted as a single column, and the cost. Order by descending cost, and do not use any subqueries. 


```python
ps.sqldf("SELECT name as facility, fullname,  CASE WHEN bk.memid = 0 THEN slots * guestcost ELSE slots * membercost END AS totalcost  FROM df_bk AS bk INNER JOIN df_fac AS fac ON bk.facid = fac.facid INNER JOIN df_mem AS mem ON bk.memid = mem.memid WHERE starttime LIKE '2012-09-14%' AND totalcost > 30.0 ORDER BY totalcost DESC ")
```

### Q9: This time, produce the same result as in Q8, but using a subquery.

```python
ps.sqldf("SELECT name as Facility, fullname as Fullname, Cost FROM (SELECT mem.memid, mem.fullname, fac.facid, fac.name, fac.membercost, fac.guestcost, bk.starttime, bk.slots, CASE WHEN mem.memid = 0 THEN fac.guestcost * bk.slots ELSE fac.membercost * bk.slots END AS cost FROM df_bk AS bk INNER JOIN df_mem AS mem ON bk.memid = mem.memid INNER JOIN df_fac AS fac ON bk.facid = fac.facid WHERE bk.starttime LIKE '2012-09-14%') WHERE cost > 30 ORDER BY cost DESC")
```

### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
ps.sqldf("SELECT fac.name AS facility_name, SUM(CASE WHEN memid = 0 THEN slots * guestcost ELSE slots * membercost END) AS facility_revenue FROM df_bk AS bk INNER JOIN df_fac AS fac ON bk.facid = fac.facid GROUP BY fac.facid HAVING facility_revenue < 1000")
```

### Q12: Find the facilities with their usage by member, but not guests 

```python
ps.sqldf("SELECT facid, name, COUNT(memid) as mem_usage FROM(SELECT bk.*, fac.* FROM df_bk AS bk INNER JOIN df_fac AS fac ON bk.facid = fac.facid WHERE memid <> 0) GROUP BY 1,2")
```

### Q13: Find the facilities usage by month, but not guests

```python
ps.sqldf("SELECT month, facid, name, COUNT(memid) as mem_usage FROM(SELECT *, strftime('%m', bk.starttime) AS Month FROM df_bk AS bk INNER JOIN df_fac AS fac ON bk.facid = fac.facid) WHERE memid <> 0 GROUP BY 1,2,3 ORDER BY 1,2,3")
```
